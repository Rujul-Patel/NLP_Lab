### POS Tagging

Importing NLTK and Donwloading Datasets

```
import nltk

from nltk import word_tokenize
from nltk.corpus import brown

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('brown')
nltk.download('universal_tagset')
nltk.download('averaged_perceptron_tagger')
```

#### Tokenize and tag all the words in the sentence with penn treebank tagset.

```
sentence = "Natural language processing (NLP) is a subfield of linguistics, computer science, and artificial intelligence concerned with the interactions between computers and human language, in particular how to program computers to process and analyze large amounts of natural language data."
nltk.pos_tag(word_tokenize(sentence))
```

### Splitting the training and test sentences

```
brown_tagged = brown.tagged_sents(categories='news',tagset='universal')

#Split the dataset into train-test 
i = int(len(brown_tagged)*0.2)

train_set = brown_tagged[i:]
test_set = brown_tagged[:i]
```

#### Traning, Testing and Evaluation of Unigram Tagger using Brown Corpus

```
#Training and evaluating unigram tagger
unigram_tagger = nltk.UnigramTagger(train_set)
acc = unigram_tagger.evaluate(test_set)

print("Accuracy : ",acc)
```

```
tagged_test_sentences = unigram_tagger.tag_sents([[token for token,tag in sent] for sent in test_set])
gold = [str(tag) for sentence in test_set for token,tag in sentence]
pred = [str(tag) for sentence in tagged_test_sentences for token,tag in sentence]
from sklearn import metrics
print(metrics.classification_report(gold, pred))
```

### Evaluate Bigrams

```
from nltk.tag import BigramTagger
```

```
#Training and evaluating unigram tagger
bigram_tagger = BigramTagger(train_set)
acc = bigram_tagger.evaluate(test_set)

print("Accuracy : ",acc)
```

```
tagged_test_sentences = bigram_tagger.tag_sents([[token for token,tag in sent] for sent in test_set])
gold = [str(tag) for sentence in test_set for token,tag in sentence]
pred = [str(tag) for sentence in tagged_test_sentences for token,tag in sentence]
from sklearn import metrics
print(metrics.classification_report(gold, pred))
```

### Evaluate Trigram

```
from nltk.tag import TrigramTagger
```

```
#Training and evaluating unigram tagger
trigram_tagger = TrigramTagger(train_set)
acc = trigram_tagger.evaluate(test_set)

print("Accuracy : ",acc)
```

```
tagged_test_sentences = trigram_tagger.tag_sents([[token for token,tag in sent] for sent in test_set])
gold = [str(tag) for sentence in test_set for token,tag in sentence]
pred = [str(tag) for sentence in tagged_test_sentences for token,tag in sentence]
from sklearn import metrics
print(metrics.classification_report(gold, pred))
```

### Combining ngram Taggers to achieve higher accuracy

```
from nltk.tag import DefaultTagger
```

```
combine_tagger = DefaultTagger('NN')

tagger = [nltk.UnigramTagger,BigramTagger,TrigramTagger]
for cls in tagger:
  combine_tagger = cls(train_set,backoff=combine_tagger)

combine_tagger.evaluate(test_set)
```

```
tagged_test_sentences = combine_tagger.tag_sents([[token for token,tag in sent] for sent in test_set])
gold = [str(tag) for sentence in test_set for token,tag in sentence]
pred = [str(tag) for sentence in tagged_test_sentences for token,tag in sentence]
from sklearn import metrics
print(metrics.classification_report(gold, pred))
```

### Brill Tagger for higher accuracy

```
from nltk.tag import brill, brill_trainer 
  
def train_brill_tagger(initial_tagger, train_sents, **kwargs): 
    templates = [ 
            brill.Template(brill.Pos([-1])), 
            brill.Template(brill.Pos([1])), 
            brill.Template(brill.Pos([-2])), 
            brill.Template(brill.Pos([2])), 
            brill.Template(brill.Pos([-2, -1])), 
            brill.Template(brill.Pos([1, 2])), 
            brill.Template(brill.Pos([-3, -2, -1])), 
            brill.Template(brill.Pos([1, 2, 3])), 
            brill.Template(brill.Pos([-1]), brill.Pos([1])), 
            brill.Template(brill.Word([-1])), 
            brill.Template(brill.Word([1])), 
            brill.Template(brill.Word([-2])), 
            brill.Template(brill.Word([2])), 
            brill.Template(brill.Word([-2, -1])), 
            brill.Template(brill.Word([1, 2])), 
            brill.Template(brill.Word([-3, -2, -1])), 
            brill.Template(brill.Word([1, 2, 3])), 
            brill.Template(brill.Word([-1]), brill.Word([1])), 
            ] 
      
    # Using BrillTaggerTrainer to train  
    trainer = brill_trainer.BrillTaggerTrainer( 
            initial_tagger, templates, deterministic = True) 
      
    return trainer.train(train_sents, **kwargs)
```

```
#Accuracy of Combined Tagger
print('Accuracy of Combined Tagger : ',combine_tagger.evaluate(test_set))

brill_tag = train_brill_tagger(combine_tagger,train_set)
print('Accuracy of Brill Tagger : ',brill_tag.evaluate(test_set))
```
