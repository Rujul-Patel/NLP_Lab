### 5. Find the number of male and female names in the names corpus. Print the first 10 male and female names.

```
from nltk.corpus import names
nltk.download('names')
```

```
print('No of Male Names   => ',len(names.words('male.txt')))
print('No of Female Names => ',len(names.words('female.txt')))
```

```
print('First 10 Male Names ')
print(names.words('male.txt')[:10])

print('\n\n')
print('First 10 Femaale Names')
print(names.words('female.txt')[:10])
```

### Print the first 15 random combine labeled male and labeled female names from names corpus

```
male = [(male,'male') for male in names.words('male.txt')]
female = [(female,'female') for female in names.words('female.txt')]
```

```
import random

combined = male + female
random.shuffle(combined)

print(combined[:15])
```
